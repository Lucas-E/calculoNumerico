# Use o método de Euler para resolver o PVI

$$
\left\{\begin{matrix}
\frac{dy}{dx} = -1.2y + 7e^{-0.3x} &  & \\ 
y(0) = 3&  &
\end{matrix}\right.
$$

```python
def derivada(t,h_acumulado):
    return (-1.2*t)+(7*(math.e**(-0.3*x)))
```

```python
def pvi(v,h,h_acumulado,func,ciclo):
    if(ciclo == 0):
        return v
    return v + h*func(v,h_acumulado)
    
```

```python
h_acumulado = 0
ciclo = 0
v = 3
valores = []
x = []
y = []

for i in range(100):
    v = pvi(v,0.5,h_acumulado,derivada, ciclo)
    valores.append([h_acumulado,v])
    x.append([h_acumulado])
    y.append([v])
    h_acumulado += 0.5
    ciclo += 1
    if(h_acumulado == 2.5):
        break
```
